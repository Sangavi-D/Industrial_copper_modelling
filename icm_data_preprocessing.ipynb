Dealing with id column of dataframe

```python
#Finding the count of unique values of id column of dataframe
df['id'].nunique()
```

```python
#Finding the rows with null values in id column of df
df[df['id'].isnull()]
```

```python
#Droping ids with NAN values
df.dropna(subset=['id'],inplace=True)
```

CONVERTING TO ITEM_DATE AND DELIVERY DATE COLUMNS TO DATETIME

```python
date_col=['item_date','delivery date']
for i in date_col:
  df[i]=pd.to_datetime(df[i],format='%Y%m%d',errors='coerce').dt.date
df.head()
```

```python
df['item_date']=pd.to_datetime(df['item_date'],format='%Y-%m-%d')
df['delivery date']=pd.to_datetime(df['delivery date'],format='%Y-%m-%d')
```

```python
df.dtypes
```

CONVERTING QUANTITY TONS,CUSTOMER,COUNTRY,APPLICATION,THICKNESS,WIDTH,SELLING PRICE COLUMNS TO NUMERICAL VALUES

```python
numeric_col = ['quantity tons','customer','country','application','thickness','width','selling_price']
for i in numeric_col:
  df[i]=df[i].apply(pd.to_numeric,errors='coerce')
```

DEALING WITH MATERIAL REFERENCE COLUMN

```python
df['material_ref'] = df['material_ref'].replace(r'^00000', None, regex=True)
```

```python
df['material_ref'] = df['material_ref'].fillna(0)
```

```python
df['material_ref'] = pd.Categorical(df['material_ref'])
```

```python
df_numeric = df.select_dtypes(include='number')
correlation_matrix = df_numeric.corr()

# Print the correlation matrix
print(correlation_matrix)
```

```python
df.isna().sum()
```

```python
df.head()
```

DEALING WITH NEGATIVE VALUES IN QUANTITY TONS AND SELLING_PRICE COLUMNS

```python
df['quantity tons']=df['quantity tons'].apply(lambda x:np.nan if x<=0 else x)
df['selling_price']=df['selling_price'].apply(lambda x:np.nan if x<=0 else x)
```

FILLING NULL VALUES

```python
df['thickness'] = df['thickness'].fillna(df['thickness'].mean())
```

```python
df['quantity tons'] = df['quantity tons'].fillna(df['quantity tons'].median())
df['selling_price'] = df['selling_price'].fillna(df['selling_price'].median())
```

```python
df['item_date'] = df['item_date'].fillna(method='ffill')
df['delivery date'] = df['delivery date'].fillna(method='ffill')
```

```python
df['status'].fillna(df['status'].mode().iloc[0],inplace=True)
df['customer'].fillna(df['customer'].mode().iloc[0],inplace=True)
df['country'].fillna(df['country'].mode().iloc[0],inplace=True)
df['application'].fillna(df['application'].mode().iloc[0],inplace=True)
```

```python
df['country']=df['country'].astype(int)
df['customer']=df['customer'].astype(int)
df['selling_price']=df['selling_price'].astype(float)
```

```python
df.dtypes
```

```python
df.describe().T
```

```python
df['item_date'] = pd.to_datetime(df['item_date'], format='%Y%m%d', errors='coerce').dt.date
df['quantity tons'] = pd.to_numeric(df['quantity tons'], errors='coerce')
df['customer'] = pd.to_numeric(df['customer'], errors='coerce')
df['country'] = pd.to_numeric(df['country'], errors='coerce')
df['application'] = pd.to_numeric(df['application'], errors='coerce')
df['thickness'] = pd.to_numeric(df['thickness'], errors='coerce')
df['width'] = pd.to_numeric(df['width'], errors='coerce')
df['material_ref'] = df['material_ref'].str.lstrip('0')
df['product_ref'] = pd.to_numeric(df['product_ref'], errors='coerce')
df['delivery date'] = pd.to_datetime(df['delivery date'], format='%Y%m%d', errors='coerce').dt.date
df['selling_price'] = pd.to_numeric(df['selling_price'], errors='coerce')
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
sns.distplot(df['quantity tons'])
plt.show()
sns.distplot(df['country'])
plt.show()
sns.distplot(df['application'])
plt.show()
sns.distplot(df['thickness'])
plt.show()
sns.distplot(df['width'])
plt.show()
sns.distplot(df['selling_price'])
plt.show()
```

```python
import numpy as np
df['selling_price_log'] = np.log(df['selling_price'])
sns.distplot(df['selling_price_log'])
plt.show()

df['quantity tons_log'] = np.log(df['quantity tons'])
sns.distplot(df['quantity tons_log'])
plt.show()

df['thickness_log'] = np.log(df['thickness'])
sns.distplot(df['thickness_log'])
plt.show()
```
